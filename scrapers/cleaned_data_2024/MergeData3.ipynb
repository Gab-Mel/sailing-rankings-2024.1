```python
import pandas as pd 
import numpy as np 
from unidecode import unidecode
```

```python
matheus = pd.read_csv("../../scraping_2024/output/matheus.csv", low_memory=False)
sillas = pd.read_csv("../../scraping_2024/output/sillas.csv", low_memory=False)
final_data_regata = pd.read_csv("final_data_regata.csv", low_memory=False)
```

```python
final_data = pd.concat([final_data_regata, matheus, sillas]).reset_index(drop=True)
final_data["ID Resultado"] = final_data.index + 1

def fix_names(value:str) -> str:
    for each_number in range(0, 10):
        value = str(value).replace(f"{each_number}", "")
    return unidecode(value).upper()

final_data["Nome Competidor"] = final_data["Nome Competidor"].apply(fix_names)
```

# Arrumando Atletas

```python
lista_de_atletas = sorted(list(final_data["Nome Competidor"].astype(str).unique()))
df_atletas = pd.DataFrame(data=lista_de_atletas, columns=["Nome Competidor"])
df_atletas["ID Competidor"] = df_atletas.index + 1
df_atletas
```

# Arrumando Competições

```python
lista_de_competicoes = sorted(list(final_data["Nome Competição"].unique()))
df_competicoes = pd.DataFrame(data=lista_de_competicoes, columns=["Nome Competição"])
df_competicoes["ID Competição"] = df_competicoes.index + 1
df_competicoes
```

# Arruamndo os ID

```python
dict_competicoes = dict()
for each_row in range(len(df_competicoes)):
    dados = df_competicoes.iloc[each_row]
    dict_competicoes[dados["Nome Competição"]] = dados["ID Competição"]

list_of_data = list()
for each_competition in final_data["Nome Competição"]:
    list_of_data.append(dict_competicoes[each_competition])

final_data["ID Competição"] = list_of_data
final_data.head()
```

```python
dict_atletas = dict()
for each_row in range(len(df_atletas)):
    dados = df_atletas.iloc[each_row]
    dict_atletas[dados["Nome Competidor"]] = dados["ID Competidor"]

list_of_data = list()
for each_atleta in final_data["Nome Competidor"]:
    list_of_data.append(dict_atletas[each_atleta])

final_data["ID Competidor"] = list_of_data
final_data.head()
```

```python
final_data["Classe Vela"].unique()
```

```python
list_of_data = list()
for each_row in range(len(final_data)):
    dados = final_data["Classe Vela"].iloc[each_row]
    list_of_data.append(dados.upper())

final_data["Classe Vela"] = list_of_data

def fix_classe(value:str) -> str:
    if value == "IQFOIL FEM. " or value == "IQFOIL FEM":
        return "IQFOIL FEM."
    elif value == "NACRA" or value == "NACRA17":
        return "NACRA 17"
    elif value == "IQFOIL MASC":
        return "IQFOIL MASC."
    else:
        return value
    
final_data["Classe Vela"] = final_data["Classe Vela"].apply(fix_classe)
final_data["Classe Vela"].unique()
```

# Nacionalidade

```python
dict_nacionalidades = dict()
for i in range(len(final_data)):
    data = final_data.iloc[i]
    dict_nacionalidades[data[2]] = data[-1]
```

```python
def arruma_nac(valor):
    return dict_nacionalidades[valor]

final_data["Nacionalidade"] = final_data["Nome Competidor"].apply(arruma_nac)
```
