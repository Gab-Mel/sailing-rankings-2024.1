# EC IQFOIL 2023 FEM

```python
url1 = "https://iqeuropeans2023greece.sailti.com/en/default/races/race-resultsall"
df1 = pd.read_csv("../scraped-data-2024/c1/European Championship 2023 IQFOIL FEM..CSV", encoding='iso-8859-1', sep=";")
df1.columns = ["Pos", "Sail", "Bow", "Crew / Club", "1Q (CR)", "2Q (CR)", "3Q (CR)", "4Q (QR)", "5Q (QR)", "6Q (CR)", "7Q (CR)", "8Q (CR)", "9F (CR)",
               "10F (CR)","11F (CR)", "12F (CR)", "13F (SL)", "14F (SL)", "15F (SL)", "16F (SL/CR)", "17F (CR)", "18F (CR)", "19F (CR)", "MR", "TOTAL", "NET"]
df1
```

```python
list_of_data = list()
for each_athlete in range(len(df1)):
    data = list(df1.iloc[each_athlete])
    for each_regata in range(4, 24):
        list_of_data.append([data[0], data[3].split("/")[0], data[1].split(" ")[0], df1.columns[each_regata], data[each_regata], data[-2], data[-1], "IQFOIL FEM.", "European Championship 2023"])

df1_auxi = pd.DataFrame(columns=["Position", "Athlete", "Nacionalidade", "Regata", "Pontuação Regata", "Total", "Net", "Classe Vela", "Nome Competição"], data=list_of_data)
df1_auxi.head()
```

```python
dataframes = [df1_auxi, df2_auxi, df3_auxi, df4_auxi, df5_auxi, df6_auxi, df7_auxi, df8_auxi, df9_auxi, df10_auxi, df11_auxi]

df_combined = pd.concat(dataframes, ignore_index=True)
df_combined.dropna(subset=['Pontuação Regata'], inplace=True)
df_combined
```

```python
def nomes(valor:str) -> str:
    return unidecode(valor).upper()

df_combined["Athlete"] = df_combined["Athlete"].apply(nomes)

def regata(valor:str) -> str:
    return valor.split(" ")[0]

df_combined["Regata"] = df_combined["Regata"].apply(regata)
```

```python
df_combined.to_csv("cluster_1_resolvido.csv")
```
