```python
import pandas as pd
```

```python
df = pd.read_excel('../scraped-data-2024/c6/26o-Centro-Sul-Americano-ILCA-6-Masc.-Fem.-Bueno-Aires-2023-1-4.xlsx')
df.head()
```

```python
# rename columns
df = df.rename(columns=
                         {'Col.': 'Posição Geral', 'NOME': 'Nome Competidor', 'Total': 'Pontuação Total'})
```

```python
df['Posição Geral'] = df.index
```

```python
#turn all the Q1, Q2, ..., F3, F4 columns into rows
df = pd.melt(df, id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett'], 
                  var_name='Flotilha', value_name='Pontuação Regata')
df['Regata'] = df['Flotilha']
```

```python
df.head()
```

```python
# rename flotilha
# from "Posição Geral" 1 to 37: "Flotilha" = "O"
# from "Posição Geral" 38 to 74: "Flotilha" = "P"

df['Flotilha'] = df['Posição Geral'].apply(lambda x: 'O' if x <= 37 else 'P')
```

```python
# se o valor em 'Pontuação Regata' NÃO for um número, 'Punição' recebe o valor de 'Pontuação Regata' e 'Pontuação Regata' recebe ''
df['Punição'] = df['Pontuação Regata'].apply(lambda x: '' if isinstance(x, int) else x)
df['Pontuação Regata'] = df['Pontuação Regata'].apply(lambda x: '' if isinstance(x, str) else x)
```

```python
df['Nome Competição'] = ['Central and South American Championship 2023'] * len(df)
df['ID Competição'] = [3] * len(df)
df['Classe Vela'] = ['Ilca 6'] * len(df)
```

```python
column_order = ['Nome Competidor', 'ID Competição', 'Classe Vela', 
                'Pontuação Regata', 'Flotilha', 'Posição Geral', 
                'Punição', 'Pontuação Total', 'Nett', 'Nome Competição', 'Regata']

df = df[column_order]
```
