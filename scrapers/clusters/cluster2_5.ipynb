```python
import pandas as pd
```

```python
df_ilca7 = pd.read_excel("../scraped-data-2024/c2_5/Ilca 7 European Championship 2021.xlsx")
df_ilca6 = pd.read_excel("../scraped-data-2024/c2_5/Ilca 6 European Championship 2021.xlsx")
```

```python
df_ilca7 = df_ilca7.drop(['Boat num', 'Sail number', 'Nat'], axis=1)
df_ilca6 = df_ilca6.drop(['Boat num', 'Sail number', 'Nat'], axis=1)
df_ilca6["Flotilha"] = "O"
```

Nome Competidor,ID Competição,Classe Vela,Pontuação Regata,Flotilha,Posição Geral,Punição,Pontuação Total,Nett,Nome Competição,Regata

```python
df_ilca7.rename(columns=
                {
                    "Competitor": "Nome Competidor",
                    "Total":"Pontuação Total",
                    "Rank": "Posição Geral"
                }, inplace=True)
df_ilca6.rename(columns=
                {
                    "Competitor": "Nome Competidor",
                    "Total":"Pontuação Total",
                    "Rank": "Posição Geral"
                }, inplace=True)
df_ilca7["Pontuação Total"] = df_ilca7["Pontuação Total"]/10
df_ilca7["Nett"] = df_ilca7["Nett"]/10

df_ilca6["Pontuação Total"] = df_ilca6["Pontuação Total"]/10
df_ilca6["Nett"] = df_ilca6["Nett"]/10
```

```python
df_ilca7 = pd.melt(df_ilca7,
    id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett', 'Flotilha'],
    value_name='Pontuação Regata',
    var_name='Regata'
)
df_ilca6 = pd.melt(df_ilca6,
    id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett', 'Flotilha'],
    value_name='Pontuação Regata',
    var_name='Regata'
)
```

```python
df_ilca7["Classe Vela"] = "Ilca 7"
df_ilca7["ID Competição"] = ""
df_ilca7["Nome Competição"] = "European Championship 2021"
df_ilca7["Punição"] = ""

df_ilca6["Classe Vela"] = "Ilca 6"
df_ilca6["ID Competição"] = ""
df_ilca6["Nome Competição"] = "European Championship 2021"
df_ilca6["Punição"] = ""

column_order = ['Nome Competidor', 'ID Competição', 'Classe Vela', 'Pontuação Regata',
                    'Flotilha', 'Posição Geral', 'Punição', 'Pontuação Total', 
                    'Nett', 'Nome Competição', 'Regata']

df_ilca7 = df_ilca7[column_order]
df_ilca6 = df_ilca6[column_order]
```
