```python
import pandas as pd
```

```python
df_470 = pd.read_excel('../source-data/OVERALL_FINAL_470.xlsx')
df_nacra = pd.read_excel('../source-data/Overall_final_NACRA.xlsx')
df_ilca_7 = pd.read_excel('../source-data/Overall_final_resuts_ILCA_7.xlsx')
df_ilca_6 = pd.read_excel('../source-data/OVERALL_FINAL_ILCA_6.xlsx')
# df_kite_men = pd.read_excel('../source-data/overall_KITE_MEN.xlsx')
# df_kite_women= pd.read_excel('../source-data/overall_KITE_WOMEN.xlsx')
```

```python
df_470 = df_470.drop(['Grp'], axis=1)
df_nacra = df_nacra.drop(['Grp'], axis=1)
df_ilca_7 = df_ilca_7.drop(['Grp', 'Club or country'], axis=1)
df_ilca_6 = df_ilca_6.drop(['Grp', 'Club or country', 'Sail'], axis=1)
# df_kite_men = df_kite_men.drop(['Sail #', 'Bow', 'Cat'], axis=1)
# df_kite_women = df_kite_women.drop(['Sail #', 'Bow', 'Cat'], axis=1)
```

```python
# rename columns
df_470 = df_470.rename(columns=
                         {'Competitor': 'Nome Competidor', 'Pts Tot\nCoef=>': 'Pontuação Total', 'Pts Ret': 'Nett', 'Pos': 'Posição Geral'})
df_nacra = df_nacra.rename(columns=
                             {'Competitor': 'Nome Competidor', 'Pts Tot\nCoef=>': 'Pontuação Total', 'Pts Ret': 'Nett', 'Pos': 'Posição Geral'})
df_ilca_7 = df_ilca_7.rename(columns=
                             {'Competitor': 'Nome Competidor', 'Pts Tot\nCoef=>': 'Pontuação Total', 'Pts Ret': 'Nett', 'Pos': 'Posição Geral'})
df_ilca_6 = df_ilca_6.rename(columns=
                             {'Competitor': 'Nome Competidor', 'Pts Tot\nCoef=>': 'Pontuação Total', 'Pts Ret': 'Nett', 'Pos': 'Posição Geral'})
# df_kite_men = df_kite_men.rename(columns=
#                                  {'Crew / Club': 'Nome Competidor', 'TOTAL': 'Pontuação Total', 'NET': 'Nett', 'Pos': 'Posição Geral', 1: 'Pontuação Regata'})
# df_kite_women = df_kite_women.rename(columns=
#                                      {'Crew / Club': 'Nome Competidor', 'TOTAL': 'Pontuação Total', 'NET': 'Nett', 'Pos': 'Posição Geral', 1: 'Pontuação Regata'})
```

```python
#turn all the 1Q, 2Q, ..., 14F, 15F, MR columns into rows
df_470 = pd.melt(df_470, id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett'], 
                  var_name='Flotilha', value_name='Pontuação Regata')
df_470['Regata'] = df_470['Flotilha']

df_nacra = pd.melt(df_nacra, id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett'],
                    var_name='Flotilha', value_name='Pontuação Regata')
df_nacra['Regata'] = df_nacra['Flotilha']


df_ilca_7 = pd.melt(df_ilca_7, id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett'],
                        var_name='Flotilha', value_name='Pontuação Regata')
df_ilca_7['Regata'] = df_ilca_7['Flotilha']


df_ilca_6 = pd.melt(df_ilca_6, id_vars=['Posição Geral', 'Nome Competidor', 'Pontuação Total', 'Nett'],
                          var_name='Flotilha', value_name='Pontuação Regata')
df_ilca_6['Regata'] = df_ilca_6['Flotilha']
```

```python
df_nacra.head()
```

```python
# define functions to rename fleet names
def rename_fleet(x):
    if x == 'MR':
        x = 'MR'
    else:
        x = 'G'
    return x

df_470['Flotilha'] = df_470['Flotilha'].apply(rename_fleet)
df_nacra['Flotilha'] = df_nacra['Flotilha'].apply(rename_fleet)
df_ilca_7['Flotilha'] = df_ilca_7['Flotilha'].apply(rename_fleet)
df_ilca_6['Flotilha'] = df_ilca_6['Flotilha'].apply(rename_fleet)
```

```python
df_470['Nome Competição'] = ['Semaine Olympique Francaise De Voile 2022'] * len(df_470)
df_470['ID Competição'] = [7] * len(df_470)
df_470['Classe Vela'] = ['470'] * len(df_470)
df_470['Punição'] = [''] * len(df_470)

df_nacra['Nome Competição'] = ['Semaine Olympique Francaise De Voile 2022'] * len(df_nacra)
df_nacra['ID Competição'] = [7] * len(df_nacra)
df_nacra['Classe Vela'] = ['Nacra'] * len(df_nacra)
df_nacra['Punição'] = [''] * len(df_nacra)

df_ilca_7['Nome Competição'] = ['Semaine Olympique Francaise De Voile 2022'] * len(df_ilca_7)
df_ilca_7['ID Competição'] = [7] * len(df_ilca_7)
df_ilca_7['Classe Vela'] = ['Ilca 7'] * len(df_ilca_7)
df_ilca_7['Punição'] = [''] * len(df_ilca_7)

df_ilca_6['Nome Competição'] = ['Semaine Olympique Francaise De Voile 2022'] * len(df_ilca_6)
df_ilca_6['ID Competição'] = [7] * len(df_ilca_6)
df_ilca_6['Classe Vela'] = ['Ilca 6'] * len(df_ilca_6)
df_ilca_6['Punição'] = [''] * len(df_ilca_6)
```

```python
# drop rows with NaN values in the column 'Pontuação Regata'
df_470 = df_470.dropna(subset=['Pontuação Regata'])
df_nacra = df_nacra.dropna(subset=['Pontuação Regata'])
df_ilca_7 = df_ilca_7.dropna(subset=['Pontuação Regata'])
df_ilca_6 = df_ilca_6.dropna(subset=['Pontuação Regata'])
# df_kite_men = df_kite_men.dropna(subset=['Pontuação Regata'])
# df_kite_women = df_kite_women.dropna(subset=['Pontuação Regata'])
```

```python
column_order = ['Nome Competidor', 'ID Competição', 'Classe Vela', 
                'Pontuação Regata', 'Flotilha', 'Posição Geral', 
                'Punição', 'Pontuação Total', 'Nett', 'Nome Competição', 'Regata']

df_470 = df_470[column_order]
df_nacra = df_nacra[column_order]
df_ilca_7 = df_ilca_7[column_order]
df_ilca_6 = df_ilca_6[column_order]
```
