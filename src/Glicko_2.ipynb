### Atletas

```python
import math

class Atleta(object):
    _tau = 0.5

    def __init__(self, nome:str, rating:float=1500, rd:float=350, vol:float=0.06) -> None:
        self.nome = nome
        self.rating = rating
        self.rd = rd
        self.vol = vol

    def display_info(self) -> None:
        print(f"Nome: {self.nome.title()} - Rating: {self.rating} - Rd: {self.rd}") 
```

```python
atletas = dict()
for cada_atleta in test_df["Nome Atleta"].unique():
    atletas[cada_atleta] = Atleta(cada_atleta)
    atletas[cada_atleta].display_info()
```

```python
def get_rating(nome:str):
    return atletas[nome].rating
```

```python
for cada_chave in test_df["ID"]:
    df_aux = test_df[test_df["ID"] == cada_chave].reset_index(drop=True)
    df_aux["Rating"] = df_aux["Nome Atleta"].apply(get_rating)
    df_aux.sort_values(by="Rating", ascending=False)


    print(df_aux)
```
