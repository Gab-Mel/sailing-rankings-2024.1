```python
import pandas as pd
```

```python
column_order = ["ID Resultado", "ID Competidor", "Nome Competidor",
    "ID Competição", "Classe Vela", "Pontuação Regata", "Descarte",
    "Flotilha", "Posição Geral", "Punição", "Pontuação Total",
    "Nett", "Nome Competição", "Regata", "Nacionalidade"]
```

```python
df1 = pd.read_csv("../kite_wc_women_over_silver.csv")
df1.head()
```

```python
df1.columns[:]
```

```python
df1 = pd.melt(df1, id_vars=df1.columns[:6], value_vars=df1.columns[6:], var_name='Regata', value_name="Pontuação Regata")
df1.head()
```

```python
column_order
```

```python
df1[['ID Resultado', 'ID Competidor', 'ID Competição']] = ''
df1['Posição Geral'] = df1['RANK'].apply(lambda x: str(x)[:-2])
df1['Nome Competidor'] = df1['HELMNAME'].apply(lambda x: str(x).split('(')[0].strip())
df1['Classe Vela'] = 'KITE FEM.'
df1['Descarte'] = df1['Pontuação Regata'].apply(lambda x: str(x).split('-')[1])
df1['Punição'] = df1['Pontuação Regata'].apply(lambda x: str(x).split('-')[2])
df1['Pontuação Regata'] = df1['Pontuação Regata'].apply(lambda x: str(x).split('-')[0])
df1['Flotilha'] = "Silver"
df1['Nett'] = df1['NETT']
df1['Nome Competição'] = "World Championship 2024"
df1['Nacionalidade'] = df1['NAT.']
df1['Pontuação Total'] = df1['TOTAL']

df1 = df1[column_order]
df1.head()
```

```python
df1.to_csv("./kite_wc_women_over_silver.csv", index=False)
```
