```python
import pandas as pd
```

# 49ER WC 2024

```python
df1 = pd.read_csv("../49er_wc.csv", encoding='iso-8859-1')

list_of_data = list()
for each_athlete in range(len(df1)):
    data = list(df1.iloc[each_athlete])
    for each_regata in range(6, 24):
        try:
            list_of_data.append([data[2], data[4].upper(), data[3].split(" ")[0], f"Q{each_regata - 5}", data[each_regata].split("-")[0], data[each_regata].split("-")[1], data[each_regata].split("-")[2], data[-2], data[-1], "49ER", "World Championship 2024", 0])
        except:
            continue
        
df1_auxi = pd.DataFrame(columns=["Position", "Athlete", "Nacionalidade", "Regata", "Pontuação Regata", "Punição", "Descarte", "Total", "Net", "Classe Vela", "Nome Competição", "Flotilha"], data=list_of_data)
df1_auxi.head()
```
